```python
import pandas as pd 

# Displaying maximum # of rows + columns in a dataframe
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

hotlineData = "/Users/haleluyagetachew/desktop/uncleaned_datasets/hotlineData.csv"
arrestDemographics = "/Users/haleluyagetachew/desktop/uncleaned_datasets/MarylandUCRArrestDemographics.csv"
regionArrests = "/Users/haleluyagetachew/desktop/uncleaned_datasets/MarylandUCRRegionArrests.csv"
unemploymentData = "/Users/haleluyagetachew/desktop/uncleaned_datasets/unemploymentbycounty.csv"

# datasets
def four_dataframes():
    df1 = pd.read_csv(hotlineData)
    df2 = pd.read_csv(arrestDemographics)
    df3 = pd.read_csv(regionArrests)
    df4 = pd.read_csv(unemploymentData)
    
    return df1, df2, df3, df4

first, second, third, fourth = four_dataframes()
```
