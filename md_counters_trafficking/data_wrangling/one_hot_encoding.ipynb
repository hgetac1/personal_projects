```python
import pandas as pd 

# Displaying maximum # of rows + columns in dataframe
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

hotlineData = "/Users/haleluyagetachew/desktop/uncleaned_datasets/hotlineData.csv"

# data sets
def single_dataframe():
    df1 = pd.read_csv(hotlineData)
    
    
    return df1

first = single_dataframe()
```

```python
# Notice 10 attributes of the hotlineData are categorical feautures (i.e values in these 10 columns are not strictly numerical)

first
```

```python
import pandas as pd

# One Hot Encoding will be used to turn categorical variables into quantitative variables.

# creating indicator variables and changing columns for clarity
dummy_variable_1 = pd.get_dummies(first['sexAndLabor']).rename(columns=lambda x: 'sexAndLabor_' + str(x))
dummy_variable_2 = pd.get_dummies(first['beggingRings']).rename(columns=lambda x: 'beggingRings_' + str(x))
dummy_variable_3 = pd.get_dummies(first['factories']).rename(columns=lambda x: 'factories_' + str(x))
dummy_variable_4 = pd.get_dummies(first['restaurantFoodService']).rename(columns=lambda x: 'restaurantFoodService_' + str(x))
dummy_variable_5 = pd.get_dummies(first['peddlingRings']).rename(columns=lambda x: 'peddlingRings_' + str(x))
dummy_variable_6 = pd.get_dummies(first['healthBeautyServices']).rename(columns=lambda x: 'healthBeautyServices_' + str(x))
dummy_variable_7 = pd.get_dummies(first['landscapingService']).rename(columns=lambda x: 'landscapingService_' + str(x))
dummy_variable_8 = pd.get_dummies(first['agriculture']).rename(columns=lambda x: 'agriculture_' + str(x))
dummy_variable_9 = pd.get_dummies(first['travelingSalesCrews']).rename(columns=lambda x: 'travelingSalesCrews_' + str(x))
dummy_variable_10 = pd.get_dummies(first['genderMinorities']).rename(columns=lambda x: 'genderMinorities_' + str(x))


# merging new data frame with original data frame
first = pd.concat([first, dummy_variable_1, dummy_variable_2, dummy_variable_3, dummy_variable_4, dummy_variable_5, dummy_variable_6, dummy_variable_7, dummy_variable_8, dummy_variable_9, dummy_variable_10], axis=1)


# merge the new df to the original df and drop the original column names that were categorical
first.drop(['sexAndLabor', 'beggingRings','factories','restaurantFoodService','peddlingRings','healthBeautyServices','landscapingService','agriculture','travelingSalesCrews','genderMinorities'], axis=1, inplace=True)
```
